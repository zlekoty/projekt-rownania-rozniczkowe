# Modelowanie systemów emerytalnych w obliczu starzejącego się społeczeństwa

W dzisiejszej Europie jesteśmy świadkami bardzo szybkiego starzenia się społeczeństw. Współczynnk dzietności spada, a średnia długość życia wzrasta. Proporcja osób na emeryturze w stosunku do odób pracujących (ODR) będzie wzrastać. 
W naszym projekcie przyjżymy się prognozom, spróbujemy zasymulować alternatywne rezultaty, a także przyjrzymy się błędom w rozwiązaniach numerycznych naszego modelu.

## Model matematyczny - rozwiązanie analityczne

Aby móc porównać rozwiązania numeryczne, najpierw musieliśmy rozwiązać problem analitycznie. W tym celu użyliśmy modelu zaproponowanego w artykule [1] $$P(t)=C(t)+W(t)+R(t)$$
gdzie:

$P(t)$ -Cała populacja

$C(t)$ -Dzieci

$W(t)$ -Osoby pracujące

$R(t)$ -Osoby na emeryturze

korzystając z następujących wskaźników:

$ b  \textrm{-współczynnik urodzeń}$,

$ \mu  \textrm{-współczynnik śmierci w grupie osób na emeryturze}$,

$ \beta \textrm{-współczynnik dzieci stających się pracownikami}$,

$ \gamma \textrm{-współczynnik pracowników przechodzących na emeryturę}$

możemy sformułować następujący układ równań:
$$\left\{\begin{array}{rcl}
\dot{C}(t)&=&bP(t)-\beta C(t)\\
\dot{W}(t)&=&\beta C(t)-\gamma W(t)\\
\dot{R}(t)&=&\gamma W(t)-\mu P(t)\\
\end{array} \right.$$
aby uprościć dalsze obliczenia przeskalujemy go

$\left\{\begin{array}{rcl}
\dot{C}(t)&=&bP(t)-\beta C(t)&(1)\\
\dot{W}(t)&=&\beta C(t)-\gamma W(t)&(2)\\
\dot{R}(t)&=&\gamma W(t)-\mu P(t)&(3)\\
P(t)&=&C(t)+W(t)+R(t)&(\bigstar)\\
\end{array} \right.$

Sumujemy $(1)$, $(2)$, $(3)$

$\dot{C}(t)+\dot{W}(t)+\dot{R}(t)=bP(t)-\beta C(t)+\beta C(t)-\gamma W(t)+\gamma W(t)-\mu P(t)$

$\dot{C}(t)+\dot{W}(t)+\dot{R}(t)=bP(t)-\mu P(t) (\bigstar \bigstar)$

liczymy pochodną $(\bigstar)$

$\dot{P}(t)=\dot{C}(t)+\dot{W}(t)+\dot{R}(t)$

Po wstawieniu do  $(\bigstar \bigstar)$ otrzymujemy

$\dot{P}(t)=P(t)(b-\mu )$

Dzielimy obustronnie przez $P(t)$

$\frac{\dot{P}(t)}{P(t)}=b-\mu (\bigstar \bigstar \bigstar)$

Definiujemy

$\left\{\begin{array}{rcl}
c(t)=\frac{C(t)}{P(t)}\\
w(t)=\frac{Wt)}{P(t)}\\
r(t)=\frac{R(t)}{P(t)}\\
\end{array} \right.$

obliczamy pochodne

$\left\{\begin{array}{rcl}
\dot{c}&=&\frac{\dot{C}P-C(t)\dot{P}}{P^2}=\frac{\dot{C}}{P}-\frac{C}{P} \cdot \frac{\dot{P}}{P}=\frac{\dot{C}}{P}-c(b-\mu )\\
\dot{w}&=&\frac{\dot{W}}{P}-w(b-\mu )\\
\dot{r}&=&\frac{\dot{R}}{P}-r(b-\mu )\\
\end{array} \right.$

$\left\{\begin{array}{rcl}
\frac{\dot{C}}{P}&=&b-\beta\frac{C}{P}&=&b-\beta c\\
\frac{\dot{W}}{P}&=&\beta\frac{C}{P}-\gamma\frac{W}{P}&=&\beta c-\gamma w\\
\frac{\dot{R}}{P}&=&\gamma\frac{W}{P}-\mu &=&\gamma w - \mu\\
\end{array} \right.$

ostatecznie

$\left\{\begin{array}{rcl}
\dot{c}&=&b-\beta c-c(b-\mu )\\
\dot{w}&=&\beta c-\gamma w-w(b-\mu )\\
\dot{r}&=&\gamma w - \mu-r(b-\mu )\\
\end{array} \right.$



Rozwiążemy teraz analitycznie otrzymany układ równań

$ (\star)
\left\{\begin{array}{rcl}
\dot{c}&=&b-c(\beta +b-\mu )\\
\dot{w}&=&\beta c-w(\gamma +b-\mu )\\
\dot{r}&=&\gamma w - \mu-r(b-\mu )\\
\end{array} \right.$

Najpierw rozwiążemy jednorodny układ równań

$\left\{\begin{array}{rcl}
\dot{c}&=&&-&c(\beta +b-\mu )\\
\dot{w}&=&\beta c&-&w(\gamma +b-\mu )\\
\dot{r}&=&\gamma w &-&r(b-\mu )\\
\end{array} \right.$

$\left[\begin{array}{ccc}
\dot{c}\\
\dot{w}\\
\dot{r}\\
\end{array}\right]
=
\left[\begin{array}{ccc}
-\beta -b +\mu &0&0\\
\beta &-\gamma -b+\mu&0\\
0&\gamma&-b+\mu\\
\end{array}\right]
\left[\begin{array}{ccc}
c\\
w\\
r\\
\end{array}\right]=Av$

wyznaczamy wrtości własne

$det(A-\lambda I)=
\begin{vmatrix}
-\beta -b +\mu -\lambda&0&0\\
\beta &-\gamma -b+\mu-\lambda&0\\
0&\gamma&-b+\mu-\lambda\\
\end{vmatrix}=
(-\beta -b +\mu -\lambda)(-\gamma -b +\mu -\lambda)(-b +\mu -\lambda)$

$\left\{\begin{array}{rcl}
\lambda _1&=&-\beta -b +\mu \\
\lambda _2&=&-\gamma -b +\mu \\
\lambda _3&=& -b +\mu 
\end{array} \right.$

szukamy wektorów własnych


Dla $\lambda _1=-\beta -b +\mu$

$
\left[\begin{array}{ccc}

\beta &-\gamma +\beta&0\\

\end{array}\right]

\end{array} \right.$

Więc

$v_1=a_1
\left[\begin{array}{ccc}

1\\
-\frac{\gamma}{\beta}
\end{array}\right]$

Dla $\lambda _2=-\gamma -b +\mu$

$
\left[\begin{array}{ccc}
\gamma -\beta&0&0\\

\end{array}\right]
\left[\begin{array}{ccc}

\left[\begin{array}{ccc}

\end{array}\right]
\Leftrightarrow
\left\{\begin{array}{rcl}

\end{array} \right.$

Więc

$v_2=a_2
\left[\begin{array}{ccc}
0\\
1\\
-1
\end{array}\right]$

Dla $\lambda _3=-b +\mu$

$
\left[\begin{array}{ccc}
-\beta&0&0\\
\beta &-\gamma&0\\
0&\gamma&0\\
\end{array}\right]
\left[\begin{array}{ccc}
c_3\\

r_3\\
\end{array}\right]=
\left[\begin{array}{ccc}
0\\
0\\
0
\end{array}\right]
\Leftrightarrow
\left\{\begin{array}{rcl}
-\beta c_3&=&0\\
\beta c_3 -\gamma w_3&=&0\\
r_3&=&r_3
\end{array} \right.$

Więc

$v_3=a_3
\left[\begin{array}{ccc}
0\\
0\\
1
\end{array}\right]$

ogólnia postac to

$z(t)=e^{\lambda _1 t}v_1+e^{\lambda _2 t}v_2+e^{\lambda _3 t}v_3$

$\left\{\begin{array}{rcl}
c&=&a_1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}\\
w&=&a_1 e^{\lambda _1 t}+a_2 e^{\lambda _2 t}\\
r&=&a_1 \frac{-\gamma}{\beta}e^{\lambda _1 t}-a_2 e^{\lambda _2 t}+a_3 e^{\lambda _3 t}
\end{array} \right.
\Rightarrow
\left\{\begin{array}{rcl}
\dot{c}&=&a_1 \lambda _1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}+\dot{a}_1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}\\
\dot{w}&=&a_1 \lambda_1 e^{\lambda _1 t}+a_2 \lambda_2 e^{\lambda _2 t}+\dot{a}_1e^{\lambda _1 t}+\dot{a}_2 e^{\lambda _2 t}\\
\dot{r}&=&a_1 \lambda_1\frac{-\gamma}{\beta}e^{\lambda _1 t}-a_2 \lambda_2 e^{\lambda _2 t}+a_3 \lambda_3 e^{\lambda _3 t} +
\dot{a}_1 \frac{-\gamma}{\beta}e^{\lambda _1 t}-\dot{a}_2 e^{\lambda _2 t}+\dot{a}_3 e^{\lambda _3 t}
\end{array} \right.
$

teraz możemy przejść do rozwiązywania niejednorodnego układu równań- postawiamy otrzymane wyniki do $(\star)$

$\left\{\begin{array}{rcl}
a_1 \lambda _1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}+\dot{a}_1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}&=&
b+\lambda_1a_1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}\\
a_1 \lambda_1 e^{\lambda _1 t}+a_2 \lambda_2 e^{\lambda _2 t}+\dot{a}_1e^{\lambda _1 t}+\dot{a}_2 e^{\lambda _2 t}&=&
a_1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}+\lambda_2(a_1 e^{\lambda _1 t}+a_2 e^{\lambda _2 t})\\
a_1 \lambda_1\frac{-\gamma}{\beta}e^{\lambda _1 t}-a_2 \lambda_2 e^{\lambda _2 t}+a_3 \lambda_3 e^{\lambda _3 t} +
\dot{a}_1 \frac{-\gamma}{\beta}e^{\lambda _1 t}-\dot{a}_2 e^{\lambda _2 t}+\dot{a}_3 e^{\lambda _3 t}&=&
\gamma(a_1 e^{\lambda _1 t}+a_2 e^{\lambda _2 t})-\mu +\lambda_3(a_1 \frac{-\gamma}{\beta}e^{\lambda _1 t}-a_2 e^{\lambda _2 t}+a_3 e^{\lambda _3 t})
\end{array} \right.
$

po uproszczeniu dostajemy

$\left\{\begin{array}{rcl}
b&=&\dot{a}_1 \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}\\
0&=&\dot{a}_1 e^{\lambda _1 t}+\dot{a}_2 e^{\lambda _2 t}\\
-\mu&=&\dot{a}_1 \frac{-\gamma}{\beta}e^{\lambda _1 t}-\dot{a}_2 e^{\lambda _2 t}+\dot{a}_3 e^{\lambda _3 t}
\end{array} \right.
$

z czego wyliczamy $\dot{a}_1$, $\dot{a}_2$, $\dot{a}_3$, a następnie je całkujemy 

$\left\{\begin{array}{rcl}
\dot{a}_1 &=& \frac{b\beta}{\gamma - \beta}e^{-\lambda _1 t}\\
\dot{a}_2 &=& -\frac{b\beta}{\gamma - \beta}e^{-\lambda _2 t}\\
\dot{a}_3 &=& (b-\mu)e^{-\lambda _3 t}
\end{array} \right.
\Leftrightarrow
\left\{\begin{array}{rcl}
a_1 &=& -\frac{b\beta}{(\gamma - \beta)\lambda_1}e^{-\lambda _1 t}&+&D_1\\
a_2 &=& \frac{b\beta}{(\gamma - \beta)\lambda_2}e^{-\lambda _2 t}&+&D_2\\
a_3 &=& e^{-\lambda _3 t}&+&D_3
\end{array} \right.
$

zatem ostatecznie

$\left\{\begin{array}{rcl}
c(t)&=&(-\frac{b\beta}{(\gamma - \beta)\lambda_1}e^{-\lambda _1 t}+D_1) \frac{\gamma-\beta}{\beta}e^{\lambda _1 t}\\
c(0)&=&-\frac{b}{\lambda_1}+D_1\frac{\gamma-\beta}{\beta}=c_0\\
D_1&=&\frac{\beta(b+c_0\beta\lambda_1}{\lambda_1 (\gamma -\beta)}\\
c(t)&=&-\frac{b}{\lambda_1}+\frac{b+c_0\lambda_1}{\lambda_1}e^{\lambda _1 t}
\end{array} \right.
$


$\left\{\begin{array}{rcl}
w(t)&=&\frac{-b\beta+(b\beta + c_0 \beta\lambda_1 )e^{\lambda _1 t}}{(\gamma - \beta)\lambda_1}+\frac{b\beta}{(\gamma - \beta)\lambda_2}+D_2 e^{\lambda _2 t}\\
w(0)&=&\frac{c_0 \beta}{\gamma -\beta}+\frac{b\beta}{\lambda_2 (\gamma -\beta)}+D_2=w_0\\
D_2&=&w_0 -\frac{c_0 \beta}{\gamma -\beta}-\frac{b\beta}{\lambda_2 (\gamma -\beta)}\\

\end{array} \right.
$

$\left\{\begin{array}{rcl}
r(t)&=&\frac{b\gamma}{(\gamma - \beta)\lambda_1}+\frac{-\gamma(b +c_0 \lambda_1}{(\gamma - \beta)\lambda_1}e^{\lambda _1 t}-\frac{b\beta}{(\gamma - \beta)\lambda_2}-\frac{w_0(\gamma - \beta)\lambda_2 -\lambda_2 c_0 \beta -b\beta}{(\gamma - \beta)\lambda_2}e^{\lambda _2 t}+1+D_3 e^{\lambda _3 t}\\

\end{array} \right.
$
